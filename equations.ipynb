## Equation Reference

---

### Subterms (shared across compartments)

#### Cytokine Signal
$$\text{Cyt}(B) = C_0 + \alpha \frac{B}{B + B_{half}}$$

> $C_0$ = baseline inflammation; $\alpha$ = max inducible cytokine

#### Macrophage & Neutrophil Killing Rates
$$k_{MA} = K_{MA} \frac{M_A \cdot B}{K_m + B} \qquad k_N = K_N \frac{N \cdot B}{K_m + B}$$

> $K_m$ = saturation constant

#### Mucociliary Clearance
$$\mu_{clear} = \gamma \cdot B \cdot \frac{M_{v,ref}}{M_v}$$

> $M_{v,ref}$ = healthy reference mucus volume; $M_v$ = current mucus volume

#### Antibiotic Tissue Concentration
$$C = \frac{a_{blood}(t) \times 1000 \times f_{partition}}{V_{blood}\text{(mL)}}$$

> $\times 1000$ converts mg → μg; $f_{partition}$ = tissue penetration fraction

#### Hill/Emax PD Model
$$l(C) = k_{max} \frac{C^n}{EC_{50}^n + C^n}$$

> $n$ = Hill coefficient

### Amoxicillin PK — Three-Compartment ODEs

$$\frac{da_g}{dt} = A_{input}(t) - (\gamma_1 + \omega_1)\, a_g$$

$$\frac{da_b}{dt} = 0.85\,\gamma_1 a_g - (\omega_2 + \gamma_2)\, a_b + \gamma_3 a_L$$

$$\frac{da_L}{dt} = \gamma_2 a_b - \gamma_3 a_L$$

> $A_{input}(t)$ = oral dosing (500 mg every 8 h); $\gamma$ = absorption/transfer rates; $\omega$ = excretion rates

### Nasal Compartment

#### Nasal Bacteria
$$\frac{dB_N}{dt} = r B_N\!\left(1 - \frac{B_N}{K_{nasal}}\right) - \gamma_N B_N\frac{M_{v,ref}}{M_{v,N}} - K_{MA}\frac{M_{A,N}\,B_N}{K_m+B_N} - K_N\frac{N_N\,B_N}{K_m+B_N} - l_N B_N$$

#### Nasal Macrophages
$$\frac{dM_{A,N}}{dt} = \sigma_{MA}(M_{A0} - M_{A,N}) - \sigma_{death} - \sigma_{used}\frac{K_{MA}\,M_{A,N}\,B_N}{K_m + B_N}$$

#### Nasal Neutrophil Recruitment
$$S_N = r_{N,max} \cdot \frac{\text{Cyt}(B_N)}{\text{Cyt}(B_N) + C_{half}} \cdot \left(1 - \frac{N_N}{N_{max}}\right)$$

#### Nasal Neutrophils
$$\frac{dN_N}{dt} = S_N - \delta_N N_N - \sigma_{used}\frac{K_N\,N_N\,B_N}{K_m + B_N}$$

#### Nasal Mucus
$$\frac{dM_{v,N}}{dt} = \sigma_{basal} + \sigma_{infect}\frac{B_N}{K_{mp} + B_N} - \gamma_N M_{v,N}$$

---

### Lung Compartment

#### Aspiration Flux (Nasal → Lung)
$$\text{asp\_flux} = \alpha \cdot B_N$$

#### Lung Bacteria
$$\frac{dB_L}{dt} = r B_L\!\left(1 - \frac{B_L}{K_{lung}}\right) + \alpha B_N - \gamma_L B_L\frac{M_{v,ref}}{M_{v,L}} - K_{MA}\frac{M_{A,L}\,B_L}{K_m+B_L} - K_N\frac{N_L\,B_L}{K_m+B_L} - l_L B_L$$

#### Lung Macrophages
$$\frac{dM_{A,L}}{dt} = \sigma_{MA}(M_{A0} - M_{A,L}) - \sigma_{death} - \sigma_{used}\frac{K_{MA}\,M_{A,L}\,B_L}{K_m + B_L}$$

#### Lung Neutrophils
$$\frac{dN_L}{dt} = S_L - \delta_N N_L - \sigma_{used}\frac{K_N\,N_L\,B_L}{K_m + B_L}$$

#### Lung Mucus
$$\frac{dM_{v,L}}{dt} = \sigma_{basal} + \sigma_{infect}\frac{B_L}{K_{mp} + B_L} - \gamma_L M_{v,L}$$
