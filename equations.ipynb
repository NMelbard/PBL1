#### Bacteria — dB/dt
$$\frac{dB}{dt} = \mu B\!\left(1 - \frac{B}{K}\right) - \frac{K_{MA}\, M_A\, B}{K_m + B} - \frac{K_N\, N\, B}{K_m + B} - k_{clear}\, B$$

#### Mucus Volume — dMv/dt
$$\frac{dM_v}{dt} = \sigma_{basal} + \sigma_{infect}\frac{B}{K_{m,mv} + B} - \gamma_{beat}\, M_v$$

#### Alveolar Macrophages — dMA/dt
$$\frac{dM_A}{dt} = \sigma_{MA}(M_{A0} - M_A) - \delta_{MA}\, M_A - k_{phago,MA}\frac{M_A\, B}{K_m + B}$$

#### Neutrophils — dN/dt
$$\frac{dN}{dt} = S_N \cdot C(B) \cdot \left(1 - \frac{N}{N_{max}}\right) - \delta_N\, N - k_{phago,N}\frac{N\, B}{K_m + B}$$
